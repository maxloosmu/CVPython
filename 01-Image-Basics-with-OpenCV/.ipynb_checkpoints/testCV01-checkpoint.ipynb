```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/00-puppy.jpg')
```

```python
img.shape
```

```python
img_bgr = img.copy()
```

```python
plt.imshow(img_bgr)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('../DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
img_gray = cv2.imread('../DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray')
```

```python
new_img = cv2.flip(img_gray,1)
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
type(new_img)
```

```python
cv2.imwrite('img_gray.jpg',new_img)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)
ax.imshow(new_img, cmap='gray')
```

```python
cv2.rectangle(new_img,pt1=(384,0),pt2=(510,128),color=(0,255,0),thickness=5)
```

```python
plt.imshow(new_img)
```

```python
cv2.rectangle(new_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=25)
```

```python
cv2.circle(img=new_img, center=(100,100), radius=50, color=(255,0,0), thickness=15)
```

```python
cv2.circle(new_img, center=(400,400), radius=50, color=(255,0,0), thickness=-1)
```

```python
cv2.line(new_img,pt1=(0,0),pt2=(511,511),color=(102, 255, 255),thickness=5)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(new_img,text='Hello',org=(10,500), fontFace=font,fontScale= 4,color=(255,255,255),thickness=5,lineType=cv2.LINE_AA)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],np.int32)
pts = vertices.reshape((-1,1,2))
cv2.polylines(new_img,[pts],isClosed=True,color=(255,0,0),thickness=5)
```

```python
cv2.imwrite('img_draw.jpg',new_img)
```
